```python
# Importing important modules

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Importing and reading the training and testing files for the model

df_x = pd.read_csv("Boston_Train.csv")
df_y = pd.read_csv("Boston_Test.csv")
```

```python
# Selecting important features from the database and removing unnecessary features.
# Setting up the target as medv(prices)

df_x_train = df_x[["crim", "rm", "dis", "rad", "lstat"]]
df_y_train = df_x["medv"]

df_x_test = df_y[["crim", "rm", "dis", "rad", "lstat"]]
df_y_test = df_y["medv"]
```

```python
# PLotting scatter plots for each feature against target variable

fig = plt.figure(figsize = (12, 7))
ax00 = fig.add_subplot(231)
ax01 = fig.add_subplot(232)
ax02 = fig.add_subplot(233)
ax10 = fig.add_subplot(234)
ax11 = fig.add_subplot(235)
ax12 = fig.add_subplot(236)

ax00.scatter(df_x["crim"], df_x["medv"])
ax01.scatter(df_x["rm"], df_x["medv"])
ax02.scatter(df_x["dis"], df_x["medv"])
ax10.scatter(df_x["rad"], df_x["medv"])
ax11.scatter(df_x["lstat"], df_x["medv"])
ax12.scatter(df_x["medv"], df_x["medv"])
```

```python
# As it is clear from the figure above this particular model can be solved by Linear Regression.

reg = linear_model.LinearRegression()
```

```python
# Performing the necessary preprocessing

reg.fit(df_x_train, df_y_train)
```

```python
df_y_pred = reg.predict(df_x_test)
```

```python
df_y_pred
```

```python
# Checking the Score for this model (Perfect Score: + or - 1)

reg.score(df_x_train, df_y_train)
```

```python
# Predicting the prices for a unknown data array 

client_data = [[0.0795, 6.579, 10.7103, 4, 5.49]] # ["crim", "rm", "dis", "rad", "lstat"]

for i, price in enumerate(reg.predict(client_data)):
    print("Predicted selling price for Client {}'s home: ${:,.2f}K".format(i+1, price))
```
